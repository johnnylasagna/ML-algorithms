```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
slope = np.random.random()*5
intercept = np.random.random()*5
count = 20
std = 3
```

```python
print(slope, intercept)
```

```python
X = np.array([i for i in range(count)])
y = np.array([slope*i+intercept for i in X])

noise = np.random.normal(0,3,count)
y = y+noise
```

```python
print(X)
print(y)
```

```python
plt.scatter(X,y)
plt.axis('equal')
plt.title('Scatter Plot for data')
```

```python
from linearRegression import LinearRegression
```

```python
model = LinearRegression(0.01,3000)
model.fit(X,y)
y_predict = [model.predict(i) for i in X]
print(model.w, model.b)
```

```python
plt.scatter(X, y, label='Actual Data')
plt.plot(X, y_predict, color='red', label='Predicted Line')
plt.axis('equal')
plt.title('Scatter Plot for data with Prediction Line')
plt.show()
```
